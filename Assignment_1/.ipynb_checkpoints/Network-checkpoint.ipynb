```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
import re
%matplotlib inline
```

```python
G = nx.DiGraph()
```

```python
df = pd.read_csv("clean_tweets.csv")
```

```python
for row in df.iterrows():
    print(re.sub("[@:.]", "", row[1][1]))
    G.add_edge(row[1][0], re.sub("[@:]", "", row[1][1]), weight = 1)
```

```python
nx.draw_networkx(G, with_labels = False)
```

```python
net_df = pd.DataFrame({"degree":G.degree(), "betweenness":nx.betweenness_centrality(G), \
                       "closeness":nx.closeness_centrality(G)})
```

```python
net_df["scaled_degree"] = net_df["degree"]/G.number_of_nodes()
```
