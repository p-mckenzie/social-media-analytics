```python
import pandas as pd
import re
```

```python
df = pd.read_csv("raw_tweets.csv")
df.shape
```

```python
df
```

```python
columns = ['post_screen_name','text_screen_name', 'content']
clean_df =  pd.DataFrame(columns=columns)
clean_df
```

```python
count = 0
for index, row in df.iterrows():
    ### Search for Mentions 
    name_lst = row['text'].split(" ")
    matching = [s for s in name_lst if "@" in s]
    if len(matching)>1:
        ### some things here
        for i in range(len(matching)):
            clean_df.loc[count+(i+1),'text_screen_name'] = matching[i]
            ### Add Screen Name
            clean_df.loc[count+(i+1),'post_screen_name'] = row['screen_name']
            ### Add Content Type
            if re.match("RT",row['text']):
                clean_df.loc[count+(i+1),'content'] = "RT"
            else:
                clean_df.loc[count+(i+1),'content'] = "Tweet"   
    elif len(matching)==0:
        clean_df.loc[count,'post_screen_name'] = row['screen_name']
        clean_df.loc[count,'text_screen_name'] = row['screen_name']
        clean_df.loc[count,'content'] = "Tweet"
    else: 
        clean_df.loc[count,'text_screen_name'] = matching[0]
        ### Add Screen Name
        clean_df.loc[count,'post_screen_name'] = row['screen_name']
        ### Add Content Type
        if re.match("RT",row['text']):
            clean_df.loc[count,'content'] = "RT"
        else:
            clean_df.loc[count,'content'] = "Tweet"
    len_df = len(clean_df)
    count += len_df
```

```python
clean_df
```

```python
def remove_weird_char(string):
    import re
    words = string.split()
    refined = []
    for i in words:
        refined.append(re.sub("[^\w]+", " ", i))  
    sentence = ' '.join([word for word in refined])
    sentence = sentence.replace(" ", "")
    return(sentence)
```

```python
clean_df["test"] = clean_df["text_screen_name"].apply(remove_weird_char)
clean_df = clean_df.reset_index()
```

```python
clean_df.loc[clean_df["test"] == '','test'] = clean_df[clean_df["test"] == '']['post_screen_name'].values
```

```python
clean_df.to_csv("clean_tweets_v2.csv",index=False)
```
